# 상품 데이터_키워드 추출

```python
import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt
```

```python
# CSV 파일 읽기
products_df = pd.read_csv('test_translated.csv')
products_df.head()
```

#### 이모티콘 제거
##### - 상품 정보에 이모티콘 삭제

```python
# 이모티콘 제거 함수 (수정된 정규 표현식 패턴 적용)
def remove_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"  # other pictographs 
                               u"\U0001F926-\U0001F937"
                               u"\U00010000-\U0010ffff"
                               u"\u200d"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
# 'Description' 컬럼의 이모티콘 제거
products_df['Description'] = products_df['Description'].apply(remove_emojis)
```

```python
def load_stopwords(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        stopwords = f.read().splitlines()
    return stopwords
```

```python
# 불용어 파일 경로
stopwords_filepath = 'stopword.txt'
stopwords = load_stopwords(stopwords_filepath)

# 한글 텍스트에서 명사, 형용사, 동사 추출하는 함수
def extract_keywords(text):
    okt = Okt()
    words = okt.pos(text, stem=True)  # stem=True: 원형 복원
    keywords = [word[0] for word in words if word[1] in ['Noun', 'Adjective'] and word[0] not in stopwords]
    return keywords

# TF-IDF 기반으로 키워드 추출하는 함수
def extract_keywords_tfidf(text, top_n=15, threshold=0.2):
    try:
        tfidf = TfidfVectorizer(tokenizer=extract_keywords)
        tfidf_matrix = tfidf.fit_transform([text])
        feature_names = tfidf.get_feature_names_out()
        scores = tfidf_matrix.toarray()[0]
        # TF-IDF 값이 threshold 이상인 단어들을 선택하여 상위 top_n 개의 키워드 추출
        sorted_indices = scores.argsort()[::-1][:top_n]
        top_keywords = [feature_names[idx] for idx in sorted_indices if scores[idx] > threshold]
        return top_keywords
    except ValueError as e:
        print(f"오류 발생: {e}. 문서를 건너뜁니다: {text}")
        return []  # 키워드를 비워둠

# Description 컬럼의 한글 텍스트에서 종합적으로 키워드 추출하여 keyword 컬럼에 추가
df['keyword'] = df['Description'].apply(lambda x: extract_keywords_tfidf(x, top_n=15, threshold=0.1))

# 결과 출력
print(df)
```
