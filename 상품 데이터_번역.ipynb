# 상품 데이터_번역 코드
##### - DeepL API 사용

```python
import pandas as pd
import requests
```

```python
# CSV 파일 읽기
products_df = pd.read_csv('test_data.csv')
products_df.head()
```

```python
import logging

# Configure logging
logging.basicConfig(level=logging.DEBUG)
```

```python
# DeepL API 키 설정
API_KEY = 'api key'
DEEPL_URL = 'https://api-free.deepl.com/v2/translate'

# Translation function with error handling
def translate_text(text, target_lang='KO'):
    try:
        if isinstance(text, list):
            translated_list = []
            for item in text:
                params = {
                    'auth_key': API_KEY,
                    'text': item,
                    'target_lang': target_lang
                }
                response = requests.post(DEEPL_URL, data=params)
                response.raise_for_status()  # Raise error for non-200 status codes
                result = response.json()
                translated_list.append(result['translations'][0]['text'])
            return translated_list
        else:
            params = {
                'auth_key': API_KEY,
                'text': text,
                'target_lang': target_lang
            }
            response = requests.post(DEEPL_URL, data=params)
            response.raise_for_status()  # Raise error for non-200 status codes
            result = response.json()
            return result['translations'][0]['text']
    except requests.exceptions.RequestException as e:
        logging.error(f"Request to DeepL API failed: {e}")
        return f"Translation failed: {str(e)}"
    except (KeyError, json.JSONDecodeError) as e:
        logging.error(f"Failed to parse DeepL API response: {e}")
        return f"Translation failed: {str(e)}"
```

```python
df= products_df
```

#### 상품명 번역

```python
# 번역할 컬럼
columns_to_translate = ['title']

# 각 컬럼을 번역하여 새로운 컬럼에 저장
for column in columns_to_translate:
    translated_column = column + '_translated'
    df[translated_column] = df[column].apply(lambda x: translate_text(x))
```
